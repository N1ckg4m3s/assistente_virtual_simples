Importações

```
import webbrowser
from gtts import gTTS
from IPython.display import Audio, display, HTML
```

Funções auxiliares

```
def Obter_Texto():
  Texto = input('Qual texto deseja falar? ')
  return Texto
```

```
def Falar(Texto):
    print(Texto)
    gtts_object = gTTS(text=Texto, lang="pt", slow=False)
    gtts_object.save("/content/gtts_response.wav")
    display(Audio("/content/gtts_response.wav", autoplay=True))

def Executar_fala(texto):
    if "abrir facebook" in texto.lower():

        display(HTML('<a href="https://www.facebook.com" target="_blank">Clique aqui para abrir o Facebook</a>'))
        Falar("Clique no link para abrir o Facebook!")

    elif "abrir youtube" in texto.lower():

        display(HTML('<a href="https://www.youtube.com" target="_blank">Clique aqui para abrir o YouTube</a>'))
        Falar("Clique no link para abrir o YouTube!")

    elif "abrir instagram" in texto.lower():

        display(HTML('<a href="https://www.instagram.com" target="_blank">Clique aqui para abrir o Instagram</a>'))
        Falar("Clique no link para abrir o Instagram!")

    elif "pesquisar por" in texto.lower():
        pesquisa = texto.lower().replace("pesquisar por", "").strip()

        display(HTML(f'<a href="https://www.google.com/search?q={pesquisa}" target="_blank">Clique aqui para pesquisar no Google</a>'))
        Falar(f"Clique no link para pesquisar por {pesquisa} no Google!")

    elif "tocar musica" in texto.lower():
        musica = texto.lower().replace("tocar musica", "").strip()

        display(HTML(f'<a href="https://www.youtube.com/results?search_query={musica}" target="_blank">Clique aqui para tocar {musica} no YouTube</a>'))
        Falar(f"Clique no link para tocar {musica} no YouTube!")

    else:
        Falar("Desculpe, não entendi o que você quer fazer.")
```

```
texto = "Olá, tudo bem? Você está no Google Colab. " \
    "Aqui, não é possível abrir páginas da web diretamente nem utilizar o microfone. " \
    "Por isso, você deve digitar os comandos que deseja executar. " \
    "Estou em fase de teste, então tenho poucas funcionalidades. " \
    "Experimente abrir o YouTube ou pedir para pesquisar algo no Google!"

Falar(texto)
```

```
## não consegui acesso ao microfone pelo colab, então a função Obter_Texto aparece um input...
Stop = False
while Stop == False:
  texto = Obter_Texto()
  if texto == 'parar':
    Stop = True
  else:
    Executar_fala(texto)
```
